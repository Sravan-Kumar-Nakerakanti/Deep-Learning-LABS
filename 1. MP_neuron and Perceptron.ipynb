```python
#MP Neuron
x1=[0,0,1,1]
x2=[0,1,0,1]
o=[0,1,1,1]
a=[0,0,0,1]
t1=1
t2=2

def mpo_pred(x1,x2,t1):
  return [1 if (a1+b1)>=t1 else 0 for a1,b1 in zip(x1,x2)]
def mpa_pred(x1,x2,t2):
  return [1 if (a2+b2)>=t2 else 0 for a2,b2 in zip(x1,x2)]
og=mpo_pred(x1,x2,t1)
print("OR Gate MP: ",og)
ag=mpa_pred(x1,x2,t2)
print("AND Gate MP: ",ag)
```

```python
#Perceptron
import random
w1=random.random()
w2=random.random()
b=random.random()

def step_fun(x):
  return 1 if x>=0 else 0
lr=0.1
epochs=75

for epoch in range(epochs):
  for i in range(len(x1)):
    p=w1*x1[i]+w2*x2[i]+b
    y_pred=step_fun(p)

    w1 +=  lr*(a[i]-y_pred)*x1[i]
    w2 += lr*(a[i]-y_pred)*x2[i]
    b += lr*(a[i]-y_pred)
def perp(x1,x2,w1,w2,b):
  return [step_fun(a1*w1+b1*w2+b) for a1,b1 in zip(x1,x2)]
pt=perp(x1,x2,w1,w2,b)
print("Perceptron: ", pt)
```
